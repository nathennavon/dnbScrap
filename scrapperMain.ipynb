```python
#Industry Results

homeURL = 'https://www.dnb.com/business-directory/company-information.janitorial-carpet-cleaning-services.us.html?page=1'

driver = webdriver.Chrome()
driver.get(homeURL)

html = driver.page_source
soup = BeautifulSoup(html, 'lxml')

links = [i.get('href') for i in soup.find_all('a')]

driver.close()

linksClean = [] 
for i in links: 
    if i != None : 
        linksClean.append(i)
        
```

```python
def fileStream(fileName,lineAdded):
    with open(fileName, 'a+') as f: 
    # Move read cursor to the start of file.
        f.seek(0)
    # If file is not empty then append '\n'
        data = f.read(100)
        if len(data) > 0 :
            f.write("/n")
    # Append text at the end of file
        f.write(lineAdded)


def pageFirms(url):
    
    driver = webdriver.Chrome()
    driver.get(url)

    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
 
    """Collects links to individual company pages"""

    page_links = [i.get('href') for i in soup.find_all('a')]
    
    driver.close()

    page_linksClean = [] 
    for i in page_links: 
        if i != None : 
            page_linksClean.append(i)
            
    for i in page_linksClean:
        if i.startswith("/business-directory/company-profiles."):
            line = 'https://www.dnb.com/' + i
            fileStream('firmLinks.txt', line)
        
#     profileLinks = []
#     for i in page_linksClean:
#         if i.startswith("/business-directory/company-profiles."):
#             profileLinks.append('https://www.dnb.com/' + i)

#     return(profileLinks)   
```

```python
"""returns a list of 50 company profile page urls"""

def profileLink_50(URL):
    driver = webdriver.Chrome()
    driver.get(URL)
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    driver.close()

    page_links = [i.get('href') for i in soup.find_all('a')]

    page_linksClean = [] 
    for i in page_links: 
        if i != None : 
            page_linksClean.append(i)
            
    profileLinks = []
    for i in page_linksClean:
        if i.startswith("/business-directory/company-profiles."):
            profileLinks.append('https://www.dnb.com/' + i)
            
            
    return(profileLinks)
```
